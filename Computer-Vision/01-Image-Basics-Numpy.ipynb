# Image Reading

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
pic = Image.open('../meta/00-puppy.jpg')
print(type(pic))
pic_arr = np.asarray(pic)
print(type(pic_arr))
```

**OR**

```python
pic_arr = plt.imread('../meta/00-puppy.jpg')
print(type(pic_arr))
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

# Individual Color Channels

```python
plt.imshow(pic_arr[:, :, 0], cmap='gray')
```

```python
plt.imshow(pic_arr[:, :, 1], cmap='gray')
```

```python
plt.imshow(pic_arr[:, :, 2], cmap='gray')
```

# Removing Contributions of Color Channels

```python
pic_red = pic_arr.copy()
pic_green = pic_arr.copy()
pic_blue = pic_arr.copy()
```

```python
pic_red[:, :, 1:] = 0

plt.imshow(pic_red)
```

```python
pic_green[:, :, 0] = 0
pic_green[:, :, 2] = 0
plt.imshow(pic_green)
```

```python
pic_blue[:, :, :2] = 0

plt.imshow(pic_blue)
```
