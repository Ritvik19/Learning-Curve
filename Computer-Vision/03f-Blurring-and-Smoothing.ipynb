```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('../meta/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img, cmap='gray')   
```

# Mean Filters

**MANUAL FILTERS**

```python
def convolve(image, kernel):
    return cv2.filter2D(image, -1, kernel)
```

```python
kernels = {
    '3x3': np.ones(shape=(3,3), dtype=np.float32)/9,
    '5x5': np.ones(shape=(5,5), dtype=np.float32)/25,
    '7x7': np.ones(shape=(7,7), dtype=np.float32)/48
}
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
for i, k in enumerate(kernels):
    filtered_img = convolve(img, kernels[k])
    ax[i].imshow(filtered_img)
    ax[i].set_title(k)
    ax[i].set_axis_off()
```

**CV2 FILTERS**

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
for i, k in enumerate([3,5,7]):
    filtered_img = cv2.blur(img, (k, k))
    ax[i].imshow(filtered_img)
    ax[i].set_title(f'{k}x{k}')
    ax[i].set_axis_off()    
```

# Gaussian Filter

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
for i, k in enumerate([3,5,7]):
    filtered_img = cv2.GaussianBlur(img, (k, k), 10)
    ax[i].imshow(filtered_img)
    ax[i].set_title(f'{k}x{k}')
    ax[i].set_axis_off()        
```

# Median Blur


```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
for i, k in enumerate([3,5,7]):
    filtered_img = cv2.medianBlur(img, k)
    ax[i].imshow(filtered_img)
    ax[i].set_title(f'{k}x{k}')
    ax[i].set_axis_off()            
```

# Bilateral Filtering

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
for i, k in enumerate([3,5,7]):
    filtered_img = cv2.bilateralFilter(img, k, 75, 75)
    ax[i].imshow(filtered_img)
    ax[i].set_title(f'{k}x{k}')
    ax[i].set_axis_off()        
```
