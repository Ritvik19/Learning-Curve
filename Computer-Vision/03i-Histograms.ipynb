```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Histogram of complete Image

```python
img = cv2.imread('../meta/bricks.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Image Histogram')
plt.show()
```

# Histogram of Masked Image

```python
img = cv2.imread('../meta/bricks.jpg')
mask = np.zeros(img.shape[:2], np.uint8)
mask[300:400, 100:400] = 255
plt.imshow(mask,cmap='gray')
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],mask,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Image Histogram')
plt.show()
```

# Histogram Equalization

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

## Grayscale

```python
img = cv2.imread('../meta/dog_backpack.png',0)
```

```python
display(img,cmap='gray')
```

```python
hist_values = cv2.calcHist([img],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_values)
```

```python
eq_img = cv2.equalizeHist(img)
display(eq_img,cmap='gray')
```

```python
hist_values = cv2.calcHist([eq_img],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_values)
```

## Color Images

```python
color_img = cv2.imread('../meta/dog_backpack.png')
show_img = cv2.cvtColor(color_img,cv2.COLOR_BGR2RGB)
# Convert to HSV colorspace to display
hsv = cv2.cvtColor(color_img, cv2.COLOR_BGR2HSV)
display(show_img)
```

```python
# Grab V channel
hsv[:,:,2] = cv2.equalizeHist(hsv[:,:,2])
```

```python
eq_color_img = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
display(eq_color_img)
```
