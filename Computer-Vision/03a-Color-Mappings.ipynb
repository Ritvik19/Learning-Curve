# Colorspaces

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../meta/00-puppy.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img)
```

```python
img = cv2.imread('../meta/00-puppy.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img)
```
