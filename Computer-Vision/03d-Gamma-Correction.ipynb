```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def load_img():
    img = cv2.imread('../meta/bricks.jpg').astype(np.float32) / 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
img = load_img()
display_img(img)
```

# Gamma Correction : Practical Effect of Increasing Brightness

```python
img = load_img()
gamma = 1/4
effected_image = np.power(img, gamma)
display_img(effected_image)
```

```python
img = load_img()
gamma = 2
effected_image = np.power(img, gamma)
display_img(effected_image)
```
