```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('../meta/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img, cmap='gray')
```

```python
def convolve(image, kernel):
    return cv2.filter2D(image, -1, kernel)
```

```python
kernels = {
    'identity': np.array([
        [0, 0, 0],
        [0, 1, 0],
        [0, 0, 0]
    ]),
    'blur': np.array([
        [0.0625, 0.125, 0.0625],
        [0.125, 0.25, 0.125],
        [0.0625, 0.125, 0.0625],
    ]),
    'top_sobel': np.array([
        [1, 2, 1],
        [0, 0, 0],
        [-1, -2, -1]
    ]),
    'bottom_sobel': np.array([
        [-1, -2, -1],
        [0, 0, 0],
        [1, 2, 1]
    ]),
    'left_sobel': np.array([
        [1, 0, -1],
        [2, 0, -2],
        [1, 0, -1]
    ]),
    'right_sobel': np.array([
        [-1, 0, 1],
        [-2, 0, 2],
        [-1, 0, 1]
    ]),
    'outline': np.array([
        [-1, -1, -1],
        [-1, 8, -1],
        [-1, -1, -1]
    ]),
    'sharpen': np.array([
        [0, -1, 0],
        [-1, 5, -1],
        [0, -1, 0]
    ]),
    'emboss1': np.array([
        [-2, -1, 0],
        [-1, 1, 1],
        [0, 1, 2]
    ]),
    'emboss2': np.array([
        [2, 1, 0],
        [1, 1, -1],
        [0, -1, -2]
    ]),
    'emboss3': np.array([
        [0, -1, -2],
        [1, 1, -1],
        [2, 1, 0]
    ]),
    'emboss4': np.array([
        [0, 1, 2],
        [-1, 1, 1],
        [-2, -1, 0]
    ]),
}
```

```python
fig, ax = plt.subplots(4, 3, figsize=(20, 15))
for i, k in enumerate(kernels):
    filtered_img = convolve(img, kernels[k])
    ax[i//3][i%3].imshow(filtered_img)
    ax[i//3][i%3].set_title(k)
    ax[i//3][i%3].set_axis_off()
```
