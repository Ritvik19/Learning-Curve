```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

from ipywidgets import interact
```

```python
img = cv2.imread('../meta/bricks.jpg').astype(np.float32)/255
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
@interact
def gamma_correction(gamma = 1.0):
    effected_img = np.power(img, gamma)
    plt.imshow(effected_img)
```
