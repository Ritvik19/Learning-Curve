```python
import cv2 
import numpy as np  

import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread('../meta/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
real_chess = cv2.imread('../meta/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess,cmap='gray')
```

```python
mat_chess = cv2.imread('../meta/chessboard_mat.jpg')
mat_chess = cv2.cvtColor(mat_chess,cv2.COLOR_BGR2RGB)
plt.imshow(mat_chess)
```

```python
gray_mat_chess = cv2.cvtColor(mat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_mat_chess,cmap='gray')
```

# Harris Corner Detection

```python
gray = np.float32(gray_flat_chess)
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
dst = cv2.dilate(dst,None)
chess = flat_chess.copy()
chess[dst>0.01*dst.max()]=[255,0,0]
plt.imshow(chess)
```

```python
gray = np.float32(gray_real_chess)
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
dst = cv2.dilate(dst,None)
chess = real_chess.copy()
chess[dst>0.01*dst.max()]=[255,0,0]
plt.imshow(chess)
```

```python
gray = np.float32(gray_mat_chess)
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
dst = cv2.dilate(dst,None)
chess = mat_chess.copy()
chess[dst>0.001*dst.max()]=[255,0,0]
plt.imshow(chess)
```

# Shi Tomasi

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,64,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess,(x,y),3,255,-1)

plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess,80,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess,(x,y),3,255,-1)

plt.imshow(real_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_mat_chess,80,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(mat_chess,(x,y),3,255,-1)

plt.imshow(mat_chess)
```
